```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing the Dataset

```python
dataset = pd.read_csv('Melbourne_House_Data.csv')
dataset.dropna(inplace=True)
dataset.head(15)
```

Label Encoding "RegionName" 

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['Regionlabel'] = le.fit_transform(dataset['Regionname'])
dataset.drop(columns=['Regionname'], inplace=True)
dataset.head(7)
```

Applying Target Encoding

```python
!pip install category_encoders
```

```python
import category_encoders as ce
encoder = ce.TargetEncoder(cols='Suburb')

dataset['Suburb_Target_Price'] = encoder.fit_transform(dataset['Suburb'], dataset['Price'] )
dataset.drop(columns=['Suburb', 'Price'], inplace=True)
dataset.head(7)
```

Applying Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

dataset['Council_Area_Encoded'] = encoder.fit_transform(dataset[['CouncilArea']])
dataset['Type_Encoded'] = encoder.fit_transform(dataset[['Type']])
dataset['Method_Encoded'] = encoder.fit_transform(dataset[['Method']])

dataset.drop(columns=['CouncilArea','Type', 'Method'], inplace=True)
dataset.head(7)
```

Other Categorical Transformations

```python
dataset['Bedroom-to-bathroom Ratio'] = dataset['Bedroom2']/dataset['Bathroom']
dataset['Land-to-building Ratio'] = dataset['BuildingArea']/dataset['Landsize']
dataset['Age of Building'] = 2024 - dataset['YearBuilt']

dataset.drop(columns=['Bedroom2','Bathroom', 'BuildingArea', 'YearBuilt', 'Landsize','Date'], inplace=True)
dataset.head(7)
```

Applying Binning on Distance

```python
bin_edges = [0, 5, 15, float('inf')]
bin_labels = ['Near', 'Medium', 'Far']

dataset['Distance_Category'] = pd.cut(dataset['Distance'], bins=bin_edges, labels=bin_labels, right=True)
dataset.drop(columns=['Distance', 'Postcode'], inplace=True)
dataset.head(10)
```

Apply Log transform

```python
dataset['Age of Building Log'] = np.log(dataset["Age of Building"])
```

Final Transformed Dataset

```python
dataset.head(25)
```

```python
dataset.to_csv('modified_Housing_Data.csv', index=False)
```
